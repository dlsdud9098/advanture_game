```python
from ai.god_brain import God
from transformers import pipeline

# system_instruction = """
# 당신은 이 세계의 신 김샐프입니다. 당신은 인간을 소환해 이 세계를 구원해야 합니다.
# 현재 소환된 플레이어 정보:
# 이름: ㅁㄴㅇㄹ, 레벨: 1, 직업: 전사, 힘: 20, 민첩: 13, 지능: 15, 운: 7
# 현재 플레이어와 대화를 통해 그들의 직업과 스탯을 설정하세요.

# """

god = God()
session1 = god.create_session()

summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0)
summarized_instruction = summarizer(
    system_instruction, max_length=1024, min_length=100, do_sample=False
)[0]["summary_text"]

print(summarized_instruction)
for _ in range(5):
    input_text = input("")
    rp = god.Chat(
        session_id=session1, instruction=summarized_instruction, input_text=input_text
    )
    print(rp)
```
